```python
# Initialize the lists to store the extracted data
year = []
brand = []
model = []
old_distance = []
fuel_type = []
transmission_type = []
emi_amount = []
area = []
prices = []

# Your dictionary of URLs
dictionary = {
    'Bangluru': [
        'https://www.cars24.com/buy-used-cars-bangalore/?listingSource=ViewAllCars&storeCityId=4709',
        'https://www.cars24.com/buy-used-cars-bangalore/?sort=lhl&serveWarrantyCount=true&listingSource=ViewAllCars&storeCityId=4709',
        'https://www.cars24.com/buy-used-cars-bangalore/?sort=plh&serveWarrantyCount=true&listingSource=ViewAllCars&storeCityId=4709',
        'https://www.cars24.com/buy-used-cars-bangalore/?sort=ohl&serveWarrantyCount=true&listingSource=ViewAllCars&storeCityId=4709',
        'https://www.cars24.com/buy-used-cars-bangalore/?sort=alh&serveWarrantyCount=true&listingSource=ViewAllCars&storeCityId=4709',
        'https://www.cars24.com/buy-used-cars-bangalore/?sort=ahl&serveWarrantyCount=true&listingSource=ViewAllCars&storeCityId=4709',
        'https://www.cars24.com/buy-used-hyundai-cars-bangalore/?f=color%3Ain%3Asilver&sort=ahl&serveWarrantyCount=true&listingSource=TabFilter&storeCityId=4709'
    ],
    'Mumbai': [
        'https://www.cars24.com/buy-used-cars-mumbai/?sort=ahl&serveWarrantyCount=true&storeCityId=2378',
        'https://www.cars24.com/buy-used-cars-mumbai/?sort=bestmatch&serveWarrantyCount=true&storeCityId=2378',
        'https://www.cars24.com/buy-used-cars-mumbai/?sort=lhl&serveWarrantyCount=true&storeCityId=2378',
        'https://www.cars24.com/buy-used-cars-mumbai/?sort=plh&serveWarrantyCount=true&storeCityId=2378',
        'https://www.cars24.com/buy-used-cars-mumbai/?sort=phl&serveWarrantyCount=true&storeCityId=2378',
        'https://www.cars24.com/buy-used-cars-mumbai/?sort=olh&serveWarrantyCount=true&storeCityId=2378',
        'https://www.cars24.com/buy-used-cars-mumbai/?sort=olh&serveWarrantyCount=true&storeCityId=2378',
        'https://www.cars24.com/buy-used-cars-mumbai/?sort=alh&serveWarrantyCount=true&storeCityId=2378',
        'https://www.cars24.com/buy-used-cars-mumbai/?sort=ahl&serveWarrantyCount=true&storeCityId=2378'],
    'Hyderabad' : ['https://www.cars24.com/buy-used-cars-hyderabad/?sort=ahl&serveWarrantyCount=true&storeCityId=3686',
                  'https://www.cars24.com/buy-used-cars-hyderabad/?sort=alh&serveWarrantyCount=true&storeCityId=3686',
                  'https://www.cars24.com/buy-used-cars-hyderabad/?sort=ohl&serveWarrantyCount=true&storeCityId=3686',
                  'https://www.cars24.com/buy-used-cars-hyderabad/?sort=olh&serveWarrantyCount=true&storeCityId=3686',
                  'https://www.cars24.com/buy-used-cars-hyderabad/?sort=phl&serveWarrantyCount=true&storeCityId=3686',
                  'https://www.cars24.com/buy-used-cars-hyderabad/?sort=plh&serveWarrantyCount=true&storeCityId=3686',
                  'https://www.cars24.com/buy-used-cars-hyderabad/?sort=lhl&serveWarrantyCount=true&storeCityId=3686',
                  'https://www.cars24.com/buy-used-cars-hyderabad/?sort=bestmatch&serveWarrantyCount=true&storeCityId=3686'
                  ],
    'Pune':['https://www.cars24.com/buy-used-cars-pune/?sort=bestmatch&serveWarrantyCount=true&storeCityId=2423',
           'https://www.cars24.com/buy-used-cars-pune/?sort=lhl&serveWarrantyCount=true&storeCityId=2423',
           'https://www.cars24.com/buy-used-cars-pune/?sort=plh&serveWarrantyCount=true&storeCityId=2423',
           'https://www.cars24.com/buy-used-cars-pune/?sort=phl&serveWarrantyCount=true&storeCityId=2423',
           'https://www.cars24.com/buy-used-cars-pune/?sort=phl&serveWarrantyCount=true&storeCityId=2423',
           'https://www.cars24.com/buy-used-cars-pune/?sort=ohl&serveWarrantyCount=true&storeCityId=2423',
           'https://www.cars24.com/buy-used-cars-pune/?sort=ohl&serveWarrantyCount=true&storeCityId=2423',
           'https://www.cars24.com/buy-used-cars-pune/?sort=ahl&serveWarrantyCount=true&storeCityId=2423'],
    'Chennai':['https://www.cars24.com/buy-used-cars-chennai/?sort=ahl&serveWarrantyCount=true&storeCityId=5732',
              'https://www.cars24.com/buy-used-cars-chennai/?sort=alh&serveWarrantyCount=true&storeCityId=5732',
              'https://www.cars24.com/buy-used-cars-chennai/?sort=alh&serveWarrantyCount=true&storeCityId=5732',
              'https://www.cars24.com/buy-used-cars-chennai/?sort=olh&serveWarrantyCount=true&storeCityId=5732',
              'https://www.cars24.com/buy-used-cars-chennai/?sort=phl&serveWarrantyCount=true&storeCityId=5732',
              'https://www.cars24.com/buy-used-cars-chennai/?sort=plh&serveWarrantyCount=true&storeCityId=5732']
}

# Iterate over each URL for each city
for city, urls in dictionary.items():
    for url in urls:
        # Fetch the page content
        page = requests.get(url)
        soup = BeautifulSoup(page.text, 'html.parser')  # Specify the parser

        # Iterate over each div you are interested in
        for i in soup.find_all('div', class_='styles_contentWrap__9oSrl'):
            t = i.find('div', class_='styles_outer__NTVth')
            text = t.text if t else ''

            # Extract year using regular expression
            years = re.findall('^\d+', text)
            year.append(years[0] if years else np.nan)

            # Extract brand using regular expression
            brands = re.findall('\d{4}(\s\w+)', text)
            brand.append(brands[0] if brands else np.nan)

            # Extract model using regular expression
            models = re.findall('\d{4}\s[a-zA-Z]+\s(\w+)', text)
            model.append(models[0] if models else np.nan)

            # Extract transmission type
            trans = re.findall('(Manual|Auto)', text)
            transmission_type.append(trans[0] if trans else np.nan)

            # Extract EMI amount
            emi = re.findall('\d+,\d+', text)
            emi_amount.append(emi[0] if emi else np.nan)

            # Extract area
            locality = i.find('div', class_='styles_ellipsis__uatjG')
            area.append(locality.text if locality else np.nan)

            # Extract price
            price = re.findall('\d+\.\d+\slakh', text)
            prices.append(price[0] if price else np.nan)

            # Extract old distance
            dist = re.findall('\d+\.\d+k', text)
            old_distance.append(dist[0] if dist else np.nan)

            # Extract fuel type
            fuels = re.findall('(Petrol|CNG|Diesel)', text)
            fuel_type.append(fuels[0] if fuels else np.nan)

# Now, you should have the same length for all lists.
# print("Year:", year)
# print("Brand:", brand)
# print("Model:", model)
# print("Fuel Type:", fuel_type)
# print("Prices:", prices)
```
